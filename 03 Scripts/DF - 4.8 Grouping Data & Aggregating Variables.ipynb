### Step 6. The team now wants to target different types of spenders in their marketing campaigns.

#### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.

#### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
# Aggregating data with transform() for mean price
ords_prods_merge['mean_prod_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking columns with labels
ords_prods_merge[['user_id', 'loyalty_flag', 'mean_prod_price']].head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_prod_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_prod_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Checking columns with new labels
ords_prods_merge[['user_id', 'mean_prod_price', 'spending_flag']].head(15)
```

### Step 7. the Instacart team wants you to determine frequent versus non-frequent customers - Order frequency flag.

#### If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.

#### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.

#### If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.

```python
# Aggregating data with transform() for mean price
ords_prods_merge['median_days_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Checking columns with labels
ords_prods_merge[['user_id', 'median_days_last_order']].head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_last_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_last_order'] > 10) & (ords_prods_merge['median_days_last_order'] <= 20), 'frequency_flag'] = 'Regular-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_last_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking rows
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
