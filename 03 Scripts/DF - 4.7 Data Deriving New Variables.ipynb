```python
# Using "if" and "then"
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range producy'
    elif (row['prices'] > 5) and (row['prices'] <=15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

#### Step 2. Change label, instead of "Busiest day" to "Busiest days" (plural) and find out the two slowest days.

```python
# For-loop clustering days_of_week based on "busy"

result_exercise = []

for value in df_merged_ords["orders_day_of_week"]:
    if value == 0 or value == 1:
        result_exercise.append("Busiest days")
    elif value == 4 or value == 3:
        result_exercise.append("Slowest days")
    else:
        result_exercise.append("Regular days")
```

```python
result_exercise
```

```python
# Adding a new column "busiest_days"
df_merged_ords['busiest_days'] = result_exercise
```

```python
# Checking the new values 
df_merged_ords['busiest_days'].value_counts(dropna = False)
```

#### Step 4. Create a new column containing these labels called “busiest_period_of_day.”

```python
# Checking hours of orders
df_merged_ords['order_hour_of_day'].value_counts(dropna = False)
```

```python
# For-loop clustering hours_of_day
result_hod = []

for value in df_merged_ords["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_hod.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result_hod.append("Average orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_hod.append("Fewest orders")
    else:
        result_hod.append("Incorrect value")
```

```python
result_hod
```

```python
# Adding the new column "busiest_period_of_day"
df_merged_ords['busiest_period_of_day'] = result_hod
```
